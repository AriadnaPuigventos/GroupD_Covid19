```python
import pandas as pd
import math
import numpy as np
```

```python
def readcsv(ruta):
      df = pd.read_csv(ruta, sep = ";")
      return df
df = readcsv(ruta="../Data/210111_air_quality_dataset.csv")
df
```

```python
df = df.replace(0, np.nan)
df
```

```python
na_ratio = ((df.isnull().sum() / len(df))*100).sort_values(ascending = False)
print(na_ratio)
```
